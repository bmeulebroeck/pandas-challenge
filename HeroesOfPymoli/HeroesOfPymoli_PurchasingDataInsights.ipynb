```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Calculate the number of unique SNs to find how many unique players
totalplayers = len(purchase_data["SN"].unique())

# Create a data frame showing the number of unique players found above and display
totalplayers_df = pd.DataFrame({"Total Unique Players":[totalplayers]})
totalplayers_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate the required values to summarize purchase data
uniqueitems = len(purchase_data["Item Name"].unique())
avgprice = purchase_data["Price"].mean()
totalpurch = len(purchase_data["Purchase ID"].unique())
totalrev = purchase_data["Price"].sum()

# Load values into a summary data frame
purchsummary_df = pd.DataFrame({"Number of Unique Items":[uniqueitems], "Average Price":[avgprice], 
                                "Number of Purchases":[totalpurch], "Total Revenue":[totalrev]})

# Use map to format the price and value columns
purchsummary_df["Average Price"] = purchsummary_df["Average Price"].astype(float).map("${:.2f}".format)
purchsummary_df["Total Revenue"] = purchsummary_df["Total Revenue"].astype(float).map("${:,.2f}".format)

purchsummary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group original DF by gender, then manipulate price to calculate...
# Total purchases by gender
genderpurchct = purchase_data.groupby("Gender").count()["Price"]
# Avg purchase price by gender
genderavgpurch = purchase_data.groupby("Gender").mean()["Price"]
# Total purchases by gender
genderpurchtot = purchase_data.groupby("Gender").sum()["Price"]
# Avg purchase total spend by gender using variables from above
genderavgtot = genderpurchtot/gendercts

gender_summary_df = pd.DataFrame({"Total Purchases":genderpurchct, "Avg Purchase Price":genderavgpurch, "Total Purchase Value":genderpurchtot, "Avg Total Spend per person":genderavgtot})

# Use map to format the price and value columns
gender_summary_df["Avg Purchase Price"] = gender_summary_df["Avg Purchase Price"].astype(float).map("${:.2f}".format)
gender_summary_df["Total Purchase Value"] = gender_summary_df["Total Purchase Value"].astype(float).map("${:.2f}".format)
gender_summary_df["Avg Total Spend per person"] = gender_summary_df["Avg Total Spend per person"].astype(float).map("${:.2f}".format)

gender_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create another copy of the data frame to do age analysis tasks
unique_snfull_df = purchase_data.copy()

# Create the bins and set labels for organizing by Age
bins = [0,9,14,19,24,29,34,39,100]
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Create the new column in the data frame with Age Bracket
pd.cut(unique_snfull_df["Age"], bins, labels=bin_labels, include_lowest=True)
unique_snfull_df["Age Bracket"] = pd.cut(unique_snfull_df["Age"], bins, labels=bin_labels, include_lowest=True)
#unique_snfull_df
```

```python
# Removing duplicate screen names to get an accurate gender breakdown. Found method on geeksforgeeks.com
unique_sn_nodupl_df = unique_snfull_df.copy()

unique_sn_nodupl_df.sort_values("SN", inplace=True)
unique_sn_nodupl_df.drop_duplicates(subset="SN", keep="first", inplace=True)
#unique_snfull_df

# Calculate total count by age bracket and percent of players by age bracket
agects = unique_sn_nodupl_df["Age Bracket"].value_counts()
agepct = agects/totalplayers

# Create the age summary dataframe and change the index order for ascending ages (found method on Stack Overflow)
age_summary_df = pd.DataFrame({"Total Count":agects, "Percent of Total Players":agepct}, index=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"])
age_summary_df.loc[["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"],:]

# Use map to format the price and value columns
age_summary_df["Percent of Total Players"] = age_summary_df["Percent of Total Players"].astype(float).map("{:.2%}".format)

# Display results
age_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Add up purchases per SN
snpurchct = purchase_data.groupby("SN").count()["Purchase ID"]
snpurchtot = purchase_data.groupby("SN").sum()["Price"]
snavgpurch = snpurchtot/snpurchct

sn_purchsummary_df = pd.DataFrame({"Purchase Count":snpurchct, "Avg Purchase Price":snavgpurch, "Total Spend":snpurchtot})

# Sort the DF highest to lowest
sn_purchsumsort_df = sn_purchsummary_df.sort_values("Total Spend", ascending=False)

# Use map to format the price and value columns
sn_purchsumsort_df["Avg Purchase Price"] = sn_purchsumsort_df["Avg Purchase Price"].astype(float).map("${:.2f}".format)
sn_purchsumsort_df["Total Spend"] = sn_purchsumsort_df["Total Spend"].astype(float).map("${:.2f}".format)

# Display the top 5
sn_purchsumsort_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a dataframe for the item purchase analysis
item_data_df = purchase_data[["Item ID", "Item Name", "Price"]]

# Calculate how many times each item was purchased
itempurchct = item_data_df.groupby(["Item ID","Item Name"]).count()["Price"]
# Total purchase value per item
itempurchtot = item_data_df.groupby(["Item ID","Item Name"]).sum()["Price"]
# Price per item
itemprice = item_data_df.groupby(["Item ID","Item Name"]).mean()["Price"]


# Load these calcs into the dataframe and sort by the Purchase Count column
item_purchase_df = pd.DataFrame({"Purchase Count":itempurchct, "Item Price":itemprice, "Total Purchase Value":itempurchtot})

# Sort by the Purchase Count Column
item_purchsort_df = item_purchase_df.sort_values("Purchase Count", ascending=False)

# Use map to format the price and value columns
item_purchsort_df["Item Price"] = item_purchsort_df["Item Price"].astype(float).map("${:.2f}".format)
item_purchsort_df["Total Purchase Value"] = item_purchsort_df["Total Purchase Value"].astype(float).map("${:.2f}".format)

# Display the top 5
item_purchsort_df.head(5)
```
